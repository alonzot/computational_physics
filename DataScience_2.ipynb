<h1> Data Modeling and Fitting </h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

def matplot_it(x,y, x_label, y_label, title):
        #define plot size in inches (width, height) & resolution(DPI)
        fig = plt.figure(figsize=(5, 5), dpi=150)

        #plot data
        plt.plot(x,y, linestyle="dashed", marker=".", color="green")

        #labels
        plt.xlabel(x_label)
        plt.ylabel(y_label)

        #title
        plt.title(title)

        #adjust plot
        plt.subplots_adjust(left=0.18)

        #show plot
        plt.show()

# Creating a function to model and create data
def func(x, a, b):
    return a * x + b

def main():
    # Generating clean data
    x = np.linspace(0, 10, 100)
    y = func(x, 1, 2)

    # Adding noise to the data
    yn = y + 0.9 * np.random.normal(size=len(x))

    # Executing curve_fit on noisy data
    popt, pcov = curve_fit(func, x, yn)

    # popt returns the best fit values for parameters of
    # the given model (func).

    print(popt)
    # call the function with the fitted parameters
    yfit=func(x, popt[0], popt[1])
    matplot_it(x,yn, 'X', 'Y', 'Raw Data')
    matplot_it(x,y, 'X', 'Y', 'Y= x +2')
    matplot_it(x,yfit, 'X', 'Y', 'Linear Regression Fit')
   
if __name__ == '__main__':
    main()
```

<h1 style="color:blue;text-align:center;"> Non-Linear Data Modeling and Fitting </h1>
<p style="color:green">The code below uses the Numpy and SciPy modules to perform a non-linear regression analysis on data with a Gaussian profile with random noise. .</p>

```python
%matplotlib inline
#from cycler import cycler
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
# Creating a function to model and create data
def func(x, a, b, c):
    return a*np.exp(-(x-b)**2/(2*c**2))

# Generating clean data
x = np.linspace(0, 10, 100)
y = func(x, 1, 5, 2)

# Adding noise to the data
yn = y + 0.2 * np.random.normal(size=len(x))

# Executing curve_fit on noisy data
popt, pcov = curve_fit(func, x, yn)

# popt returns the best-fit values for parameters of the given model (func).
print(popt)
yfit=func(x,popt[0],popt[1],popt[2])
#matplot_it(x,y, 'X', 'Y', 'Non-Linear Regression Fit')
 #define plot size in inches (width, height) & resolution(DPI)
fig = plt.figure(figsize=(7, 7), dpi=150)
#plt.gca().set_color_cycle(['red', 'green', 'blue', 'yellow'])
#plt.rc('axes', prop_cycle=(cycler('color', ['g', 'r', 'b', 'y']) + cycler('linestyle', ['-', '--', ':', '-.'])))
#plot data
plt.scatter(x,yn, linestyle="dashed", marker=".", color="red", label = 'data')
plt.plot(x,y, linestyle="dashed", marker=".", color="green", label = 'function')
plt.plot(x,yfit, linestyle="dashed", marker=".", color="blue", label = 'best fit')
#Axis labels
plt.xlabel(' x values')
plt.ylabel(' y values')

#title
plt.title('Non-Linear Regression Analysis')
#legends
#plt.legend(['data', 'y=gaussian', 'best fit'], loc='upper left')
plt.legend(loc='upper left')
#adjust plot
plt.subplots_adjust(left=0.18)
#show plot
plt.show()

#fig, ax = plt.subplots()
#ax.plot(x, yn, 'k--', label='Model length')
#ax.plot(x, y, 'k:', label='Data length')
#ax.plot(x, yfit, 'k', label='Total message length')
# Now add the legend with some customizations.
#legend = ax.legend(loc='upper center', shadow=True)
```
